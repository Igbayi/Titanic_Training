```python
##One Hot Encoding

df = pd.get_dummies(df, columns = ["Gender"], drop_first = True)
```

```python
y = df["Salary "]
```

```python
x = df[['YearsExperience', 'Gender_Male']]
x
```

```python
#Model selection and model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size= 0.3,random_state= 1)
```

```python
y_test
```

```python
x_train = pd.DataFrame(x_train)
x_test = pd.DataFrame(x_test)
```

```python
x_test
```

```python
model = LinearRegression()   #build linear regression model (Intialistiate)
model
model.fit(x_train,y_train)   #### ML to learn
```

```python
Pred = model.predict(x_test)

Pred
```

```python
#Evaluation Metrics
from sklearn.metrics import mean_squared_error     
from sklearn import metrics

import warnings
warnings.filterwarnings("ignore")
```

```python
plt.scatter(x,y, color = "red")
plt.plot(x_test,Pred, color = "black")
plt.title("Linear Regression of Years of Exp and Salary")
plt.xlabel ("Years of Exp")
plt.ylabel ("Salary")

plt.show()
```

```python
print("MSE", mean_squared_error(y_test,Pred))
print("R squared", metrics.r2_score(y_test,Pred))
```
